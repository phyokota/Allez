```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cleaned_climbharder_data.csv")

x_col = "project_v_grade"
y_col = "max_hang_opencrimp_18mm_10s_kg"

plot_df = df[[x_col, y_col]].dropna()

plot_df.plot(
    kind="scatter",
    x=x_col,
    y=y_col,
    figsize=(8, 5),
    alpha=0.7,
)

plt.title("Project Grade vs Max Hang Open Crimp")
plt.xlabel("Project Grade")
plt.ylabel("Max Hang Open Crimp 18mm 10s (kg)")
plt.show()

df.groupby("project_v_grade")["max_hang_opencrimp_18mm_10s_kg"].median()

x_col = "project_v_grade"
y_col = "min_edge_opencrimp_mm"

plot_df = df[[x_col, y_col]].dropna()

plot_df.plot(
    kind="scatter",
    x=x_col,
    y=y_col,
    figsize=(8, 5),
    alpha=0.7,
)

plt.title("Project Grade vs Min Edge Open Crimp")
plt.xlabel("Project Grade")
plt.ylabel("Min Edge Open Crimp")
plt.show()

df.groupby("project_v_grade")["min_edge_opencrimp_mm"].median()

df.groupby("project_v_grade")["max_pullup_reps"].median()

```
