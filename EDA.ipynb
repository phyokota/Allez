```python
import pandas as pd
```

```python
df = pd.read_csv('climbharder_data.csv')

#Scanning through the 20 columns
#print(df["timestamp"].unique())
#print(df["sex"].unique())

cols = ["height_cm", "weight_kg", "arm_span_cm"]

#Cleaned cols with weird values
df[cols] = df[cols].apply(pd.to_numeric, errors="coerce")

#Removes "years" from years_climbing column and averges the range
years = df["years_climbing"].str.findall(r"\d*\.?\d+")
df["years_climbing"] = years.apply(lambda nums: pd.to_numeric(nums).mean())

#Cleaned cols max_v_grade and project_v_grade
df["max_v_grade"] = pd.to_numeric(df["max_v_grade"].str.lower().str.removeprefix("v"))
df["project_v_grade"] = pd.to_numeric(df["project_v_grade"].str.lower().str.removeprefix("v"))

#First 10 cols cleaned

#Cols 11-13 are clean

#Col 14
df["max_hang_halfcrimp_18mm_10s_kg"] = df["max_hang_halfcrimp_18mm_10s_kg"].str.lower().str.replace(r"\s*kg$", "", regex=True)
df["max_hang_halfcrimp_18mm_10s_kg"] = pd.to_numeric(df["max_hang_halfcrimp_18mm_10s_kg"], errors="coerce")
df["max_hang_halfcrimp_18mm_10s_kg"] = abs(df["max_hang_halfcrimp_18mm_10s_kg"])

#Col 15
df["max_hang_opencrimp_18mm_10s_kg"] = df["max_hang_opencrimp_18mm_10s_kg"].str.lower().str.replace(r"\s*kg$", "", regex=True)
df["max_hang_opencrimp_18mm_10s_kg"] = pd.to_numeric(df["max_hang_opencrimp_18mm_10s_kg"], errors="coerce")
df["max_hang_opencrimp_18mm_10s_kg"] = abs(df["max_hang_opencrimp_18mm_10s_kg"])

#Col 19 & 20 decided to remove, people inputed values in lbs and kg data too messy, also pushups aren't really correlated
df = df.drop(columns=["weighted_pullup_5rm_kg", "max_pushup_reps"])

#Col 16,17,18,20
cols = ["min_edge_halfcrimp_mm", "min_edge_opencrimp_mm", "max_pullup_reps"]
df[cols] = df[cols].apply(pd.to_numeric, errors="coerce")

#creating primary key for supabase
df.insert(0, "id", range(1, len(df) + 1))

#converting to cleaned csv
df.to_csv("cleaned_climbharder_data.csv", index=False)





```
